```python
from sklearn.feature_extraction.text import TfidfVectorizer

#collecting strings from documents and create a corpus
d0 = 'Geeks for geeks'
d1 = 'Geeks'
d2 = 'r2j'
string = [d0, d1, d2]
```

```python
#fecthing tf-idf values
tfidf = TfidfVectorizer()
result = tfidf.fit_transform(string)
print('\nidf values:')
for ele1, ele2 in zip(tfidf.get_feature_names_out(), tfidf.idf_):
    print(ele1, ':', ele2)
```

```python
#another method
documents = [
    "NLP is very powerful",
    "NLP helps machines understand language",
    "Understanding language is important"
]

# Create vectorizer
tfidf1 = TfidfVectorizer()

# Fit and transform documents
tfidf_matrix = tfidf1.fit_transform(documents)

# Feature names (words)
print(tfidf1.get_feature_names_out())

# TF-IDF values
print(tfidf_matrix.toarray())
```
