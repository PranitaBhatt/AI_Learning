```python
import nltk 
from nltk.tokenize import word_tokenize, sent_tokenize, RegexpTokenizer

nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
#Implementing word tokenize
text="Implementing word tokenization in NLP"
tokens=word_tokenize(text)
print(tokens)
```

```python
#Implementing sentence tokenization
sentence="Implementing word tokenization in NLP. It is an interesting Subject to work with. We will further work on TF-IDF"
tokenss=sent_tokenize(sentence)
print(tokenss)
```

```python
#Implementing Character Tokenization
characters="NLP with AIML"
char_tokenizer = RegexpTokenizer(r'.', gaps=False)  #will match every single character
tokensc = char_tokenizer.tokenize(characters)
print(tokensc)

```
