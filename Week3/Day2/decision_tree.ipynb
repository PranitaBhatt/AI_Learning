```python
# Basic Decision Tree Implementation (Classification)

import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree

# 1. Create a small dataset
# Features: [Age, Salary]
X = np.array([
    [25, 50000],
    [30, 60000],
    [35, 65000],
    [40, 70000],
    [45, 80000],
    [50, 90000]
])

# Target: Buy product (0 = No, 1 = Yes)
y = np.array([0, 0, 0, 1, 1, 1])
```

```python
# 2. Create and train the Decision Tree model
model = DecisionTreeClassifier(
    criterion="gini",   # or "entropy"
    max_depth=3         # limit tree depth to avoid overfitting
)

model.fit(X, y)
```

```python
# 3. Make a prediction
new_customer = [[42, 72000]]
prediction = model.predict(new_customer)
print("Prediction (0 = No, 1 = Yes):", prediction)
```

```python
# 4. Check training accuracy (for learning purposes)
accuracy = model.score(X, y)
print("Training Accuracy:", accuracy)
```

```python
# 5. Visualize the decision tree
plt.figure(figsize=(12, 6))
plot_tree(
    model,
    feature_names=["Age", "Salary"],
    class_names=["No", "Yes"],
    filled=True
)
plt.show()
```
