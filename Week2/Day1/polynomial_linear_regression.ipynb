```python
#Polynomial Regression is used when data points are not linearly aligned. It fits a polynomial curve to the data.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
x=6*np.random.rand(100,1)-3   #100 random numbers between 0-1 and 1 column
#6 is used for scaling which changes range from 0-1 to 0-6. It is used for To make x‑values spread wider instead of being tiny decimals
#-3 is used for shifting which changes range from 0-6 to -3 to 3. It is used to center the x‑values around zero instead of being all positive.

y=0.5*x**2+1.5*x+2+np.random.rand(100,1)  #This is a quadratic equation
# x**2 itroduces the curve
# 0.5*x**2 control curvature strength  --Bigger value → steeper curve, Smaller value → flatter curve--
# +1.5*x adds a tilt to the curve
# +2 intercept
# np.random.rand(100,1) adds noise to make the data more realistic by introducing variability.

plt.scatter(x,y,color='g') #g for green color
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomial Regression')
```

```python
polynomial_features=PolynomialFeatures(degree=2)  #degree=2 for quadratic features
x_poly=polynomial_features.fit_transform(x)  #Transforming x to include polynomial features
model=LinearRegression()  #Creating a linear regression model
model.fit(x_poly,y)  #Fitting the model to the polynomial features and target variable
y_poly_pred=model.predict(x_poly)  #Predicting the target variable using the fitted model
plt.scatter(x,y_poly_pred,color='r') #r for red color
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomial Regression Fit')
plt.show()
```
